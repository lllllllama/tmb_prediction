```python
import pandas as pd
import numpy as np
from os.path import join as j
```

```python
n_slides = 500
n_patients = 450
label_classes = {'subtype_1': 0.5, 'subtype_2': 0.3, 'subtype_3': 0.2}
save_name = 'tumor_subtyping_dummy_clean.csv'

slide_ids = ['slide_%i' %i for i in range(n_slides)]
patient_ids = np.array([i for i in range(n_slides)])
n_replace = n_slides - n_patients
patient_ids[-n_replace:] = np.random.randint(0, n_patients, (n_replace))
patient_ids = sorted(patient_ids)
patient_ids = ['patient_%i' %i for i in patient_ids]


labels = []
for name, frac in label_classes.items():
    labels.extend([name for i in range(int(n_slides * frac))])


labels = np.array(labels)[np.random.choice(np.arange(n_slides), n_slides, replace=False)]
assert len(labels) == n_slides

df = pd.DataFrame({'case_id': patient_ids, 'slide_id': slide_ids, 'label': labels})
df.to_csv(j('../dataset_csv/', save_name), index=False)
```
